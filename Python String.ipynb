```python
#String Operators
#1) Concatination (Combining two or more strings)
str1 = "Pra"
str2 = "nay"
str3 = "Reddy"
print(str1 + str2 +" " + str3)

A = "I am"
B = "a hero."
print(A +" "+ B)

X = "I am"
Y = " learning "
Z = "python."
print(X + Y + Z)

#2) Repetition
str1 = "Pra"
str2 = "nay"
str3 = "Reddy, "
result = (str1 + str2 +" " + str3)
print(result * 3)

talent1 = "cricket, "
talent2 = "kabaddi, "
talent3 = "kho-kho"
talents = talent1 + talent2 + talent3
print(talents)
print(talents * 4)

#3) Indexing
my_string = "I love myself"
print(my_string[3])

college = "Vaagdevi College of Engineering"
print(college[13])

#4) Slicing
Institute = "Teks academy"
print(Institute[5 : 12])

company = "accenture"
print(company[1 : ])
```

```python
#Formating String
#1) Old style format
company = "Revature"
location = "Hyderabad"
formatted_string = "I have been working in %s at %s for 2 years."%(company,location)
print(formatted_string)

institute = "teks academy"
years = 15
formatted_string = "There is a institute called %s since %d years."%(institute,years)
print(formatted_string)

name = "pranay"
age = 22
formatted_string = "my name is %s and i am %d years old."%(name,age)
print(formatted_string)

#2) format method
vechicle = "bike"
time = 2
formatted_string = "I have went to college on my {} within {} hours.".format(vechicle,time)
print(formatted_string)

no_of_fruits = "three"
types_of_fruits = "apples"
formatted_string = "He ate all the {} {}.".format(no_of_fruits,types_of_fruits)
print(formatted_string)

#3) String interpolation (f-string)
place = "AMB mall"
no_of_people = 200
formatted_string = f"There are {no_of_people} members inside the {place}."
print(formatted_string)

jobs = 10
competition = "lot of people"
formatted_string = f"There are {competition} for just {jobs} openings."
print(formatted_string)
```

```python
#String Method
#1) len()
Name = "Pranay Kanchanakuntla"
length = len(Name)
print(length)

Location = "My current location is Hyderabad"
length = len(Location)
print(length)

#2) lower() / upper()
my_string = "i am leaving Hyderabad"
print(my_string.upper())

name = "pranay"
uppercase = my_string.upper()
print(uppercase)

skill = "I HAVE GOOD KNOWLEDGE IN PYTHOM"
print(skill.lower())

sports = "I PLAY CRICKET WELL"
print(sports.lower())


#3) Strip()
Village = " Sriramulapally      "
print(Village.strip())

abc = "   we are good friends     "
print(abc.strip())

#4) Replace
City = "I am in Hanamkonda"
new_string = City.replace("Hanamkonda","Hydearabad")
print(new_string)

skills = "I have good idea on Python and SQL"
print(skills.replace("idea on","knowledge in"))

#5) Spilt()
about = "Life has no Cntrl+Z"
print(about.split())

location = "I have been living in Hanamkonda for 22 years"
print(location.split())

#6) join()
about = ['Life', 'has', 'no', 'Cntrl+Z']
join_str = " ".join(about)
print(join_str)

location = ['I', 'have', 'been', 'living', 'in', 'Hanamkonda', 'for', '22', 'years']
print(' '.join(location))
```
